```python
import pandas as pd
import numpy as np
import shap
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2021.csv")
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 100], labels=['<30', '30-45', '45-60', '60+'])
df['BMIGroup'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
df['BP_Chol_Risk'] = df['HighBP'] + df['HighChol']
```

```python
df.dropna(subset=['AgeGroup', 'BMIGroup'], inplace=True)
```

```python
scaler = StandardScaler()
df[['BMI', 'PhysHlth', 'MentHlth']] = scaler.fit_transform(df[['BMI', 'PhysHlth', 'MentHlth']])
```

```python
X = df.drop(columns=['Diabetes_binary', 'AgeGroup', 'BMIGroup'])
y = df['Diabetes_binary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    print(f"\n{name} Results:")
    print(classification_report(y_test, y_pred))
    print(f"ROC-AUC: {roc_auc:.4f}")
```

```python
explainer = shap.Explainer(models['XGBoost'])
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)
```

```python
streamlit_code = '''
import pandas as pd
import numpy as np
import streamlit as st
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

# Load Dataset
df = pd.read_csv("diabetes_binary_health_indicators_BRFSS2021.csv")

# Feature Engineering
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 30, 45, 60, 100], labels=['<30', '30-45', '45-60', '60+'])
df['BMIGroup'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
df['BP_Chol_Risk'] = df['HighBP'] + df['HighChol']

# Drop rows with missing values in newly created columns
df.dropna(subset=['AgeGroup', 'BMIGroup'], inplace=True)

# Normalize continuous variables
scaler = StandardScaler()
df[['BMI', 'PhysicalHealth', 'MentalHealth']] = scaler.fit_transform(df[['BMI', 'PhysicalHealth', 'MentalHealth']])

# Prepare data for modeling
X = df.drop(columns=['Diabetes_binary', 'AgeGroup', 'BMIGroup'])
y = df['Diabetes_binary']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Initialize Models
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')
}

# Evaluate models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    print(f"\\n{name} Results:")
    print(classification_report(y_test, y_pred))
    print(f"ROC-AUC: {roc_auc:.4f}")
    
# SHAP Explainability for XGBoost
explainer = shap.Explainer(models['XGBoost'])
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)

# Streamlit UI
st.title(" Diabetes Risk Prediction")

st.write("Fill in the details below to check the diabetes risk")

bmi = st.slider('BMI (Normalized)', float(X['BMI'].min()), float(X['BMI'].max()), float(X['BMI'].mean()))
phys_health = st.slider('Physical Health (Normalized)', float(X['PhysicalHealth'].min()), float(X['PhysicalHealth'].max()), float(X['PhysicalHealth'].mean()))
mental_health = st.slider('Mental Health (Normalized)', float(X['MentalHealth'].min()), float(X['MentalHealth'].max()), float(X['MentalHealth'].mean()))
high_bp = st.selectbox('High Blood Pressure', [0, 1])
high_chol = st.selectbox('High Cholesterol', [0, 1])
smokers = st.selectbox('Smoker', [0, 1])
physically_active = st.selectbox('Physically Active', [0, 1])
diabetic = st.selectbox('Previously Diagnosed Diabetes', [0, 1])
stroke = st.selectbox('Stroke', [0, 1])
heart_disease = st.selectbox('Heart Disease or Attack', [0, 1])

submit = st.button("Predict Risk")

if submit:
    input_data = pd.DataFrame({
        'BMI': [bmi],
        'PhysicalHealth': [phys_health],
        'MentalHealth': [mental_health],
        'HighBP': [high_bp],
        'HighChol': [high_chol],
        'Smoker': [smokers],
        'PhysActivity': [physically_active],
        'Diabetes_012': [diabetic],
        'Stroke': [stroke],
        'HeartDiseaseorAttack': [heart_disease],
        'BP_Chol_Risk': [high_bp + high_chol]
    })

    prediction = models['XGBoost'].predict(input_data)
    risk_score = models['XGBoost'].predict_proba(input_data)[0][1]

    if prediction[0] == 1:
        st.error(f"High risk of diabetes. Risk Score: {risk_score:.2f}")
    else:
        st.success(f"Low risk of diabetes. Risk Score: {risk_score:.2f}")
'''
```
